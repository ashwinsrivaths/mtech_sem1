# pandas
- rows - observations
- columns - features

just like dictionary, using string labels data can be accessed

in index slicing works normally but in labeled slicing, endpoint is included


```python
import pandas as pd
import numpy as np
```

```python
valList = [100,200,300,400]
indexList = ['a','b','c','d']
ps1 = pd.Series(valList, index=indexList)
print(type(ps1)) # <class 'pandas.core.series.Series'>
print(ps1.shape)
print(ps1.index)
ps1
```

```python
md1 = {'a':1, 'b':2, 'c':3, 'd':4, 'e':5}
ps2 = pd.Series(md1)
ps2.index
ps2.values
type(
ps2.values
)
```

```python
print(ps2[1:3])      # index slicing works normallw with end index not included
print(ps2['b':'c'])  # labled slicing included the end lable as well
```

```python
ps1 + ps2
ps2 + ps1
```

# Data Frames

```python
rowList = ['tata','birla', 'wipro', 'infy']
collist = ['rev', 'exp', 'hq']
vlist=[
    [100,50,'mum'],
    [200,150,'kol'],
    [300,250,'blr'],
    [400,350,'blr']
]
df1 = pd.DataFrame(vlist)
print(df1.index)
print(df1.columns)
print(df1.shape)
print(df1)
```

```python
df1 = pd.DataFrame(vlist, index=rowList, columns=collist)
print(df1.index)
print(df1.columns)
print(df1.shape)
print(df1)
```
